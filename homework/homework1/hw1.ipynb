````python
import pandas as pd
from numpy import * 
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import * 
````

````python
data = pd.read_csv("student-mat.csv", sep=";")
data.head()
````

````python
data.describe()
````

````python
containsNaN = data.isnull().values.any() 
if containsNaN == False: 
    print("No NaN!")
else:
    print("Contains NaN")
````

````python
data.dtypes
````

````python
data['romantic']
````

````python
data.sex = pd.Categorical(data.sex).codes
data.Pstatus = pd.Categorical(data.Pstatus).codes
data.head()
````

````python
twoFeatures = ['age', 'sex']
````

````python
X,y = data[twoFeatures], data.loc[:, ['G3']] # train and test with 2 parameters 
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.2)

# how to select 1 column as DF and not pandas series: https://stackoverflow.com/questions/16782323/python-pandas-keep-selected-column-as-dataframe-instead-of-series 
````

````python
clf = tree.DecisionTreeClassifier(max_leaf_nodes=3, random_state = 100)
clf = clf.fit(X_train,y_train)
````

````python
y_pred = clf.predict(X_test)
treeTwoFeatures = tree.export_text(clf,feature_names = twoFeatures)
print(treeTwoFeatures)
````

````python
print(classification_report(y_test, y_pred))
````

From the accuracy above, we can see that predicting the grade (`G3`) from solely `age` and `sex` may not produe the desired results.  

Trying with 3 features.

````python
features = ['romantic', 'internet', 'Pstatus']

data.romantic = pd.Categorical(data.romantic).codes
data.internet = pd.Categorical(data.internet).codes
````

````python
X,y = data[features], data.loc[:, ['G3']] # train and test with 2 parameters 
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.2, random_state = 100)
clf = tree.DecisionTreeClassifier(max_leaf_nodes=3, random_state = 100)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
treeThreeFeatures = tree.export_text(clf,feature_names = features)
print(treeThreeFeatures)
````

````python
print(classification_report(y_test, y_pred))
````

Somehow, this is even worse. Trying with more features regarding health and increasing `max_leaf_nodes` to 4.

````python
features = ['health', 'absences', 'freetime', 'goout']
````

````python
X,y = data[features], data.loc[:, ['G3']] # train and test with 2 parameters 
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.2)
clf = tree.DecisionTreeClassifier(max_leaf_nodes=5, random_state = 100)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
treeFourFeatures = tree.export_text(clf,feature_names = features)
print(treeFourFeatures)
````

````python
print(classification_report(y_test, y_pred))
````

Getting better.... trying with more features related to student's education.  

`schoolsup`, `famsup`, `paid`, `health`, `absences`, `failures`, and `studytime` are all related to education, (mostly) independent of the student's demographics. 

````python
data.paid = pd.Categorical(data.paid).codes
data.famsup = pd.Categorical(data.famsup).codes
features = ['health', 'absences', 'famsup', 'paid', 'failures', 'studytime']
````

````python
X,y = data[features], data.loc[:, ['G3']] # train and test with 2 parameters 
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.2)
clf = tree.DecisionTreeClassifier(max_leaf_nodes=9, random_state = 100)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
treeFiveFeatures = tree.export_text(clf,feature_names = features)
print(treeFiveFeatures)
````

````python
print(classification_report(y_test, y_pred))
````

Trying all the features previously mentioned:

````python
print(classification_report(y_test, y_pred))
````
